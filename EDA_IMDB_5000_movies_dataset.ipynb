## Insights

### Finding out the number of movies per language

```python
lang_grp = df.groupby('language')
```

```python
lang_grp.count()
# The languages with the most movies are English(3608), French(34) and followed by Spanish(23)
```

### A histogram of the number of movies produced per year

```python
df['title_year'].plot(kind = 'hist', color = 'c')
```

### Finding out the number of movies per country

```python
country_grp = df.groupby('country')
```

```python
country_grp.count()
```

### Determining the movie with the highest gross figure

```python
print("The highest grossing figure is", df.gross.max())
```

```python
gross_max = df[df.gross == df.gross.max()]
print("The movie with the highest grossing figure is", gross_max.movie_title)
```

###  Plotting a scatter plot of the gross figure vs the imdb score

```python
#df.plot(kind = "scatter", x = 'imdb_score', y = 'gross')
sns.scatterplot(data = df, x = 'imdb_score', y = 'gross', hue = 'content_rating', palette = 'deep')
```

### The relation between IMDB score vs duration, taking into account the content rating(target audience)

```python
sns.relplot(data = df, x = 'imdb_score', y = 'duration', col = 'color',  style = 'content_rating', kind = 'scatter')
```

### Visualising the relationship between each variable and the IMDB score
#### to see what are the factors that drive a high IMDB rating

```python
score = sns.PairGrid(df)
score.map(plt.scatter)
```

### Highest rated movie in each audience category(content-rating)

```python
audience_grp = df.groupby('content_rating')
```

```python
#The highest rated movie in each category based content-rating and IMDB score
imdb_audience = audience_grp.max()
imdb_audience
```
